```python
import pandas as pd

# Change the path to the location where the chopstick-effectiveness.csv file is located on your computer.
# If you get an error when running this block of code, be sure the chopstick-effectiveness.csv is located at the path on your computer.

stroop_df = pd.read_csv('stroopdata.csv')
#print stroop_df

my_stats = stroop_df.describe(percentiles=[])
print my_stats
    
plot(stroop_df['Congruent'], '.', label = 'Congruent')
plot(stroop_df['Incongruent'], '.r', label = 'Incongruent')
title('Stroop Experiment Performance by Subject')
xlabel('Subject')
ylabel('Time [seconds]')
legend(numpoints=1);
```

```python
bins = range(int(round(my_stats.min().min())), int(round(my_stats.max().max())), 1)

hist(stroop_df['Congruent'], normed=True, bins=20, range=(my_stats.min().min(), my_stats.max().max()))
title('Congruent Data')
xlabel('Task Time [s]')

#subplot(1,3,2)
hist(stroop_df['Incongruent'], normed=True, bins=20, range=(my_stats.min().min(), my_stats.max().max()), alpha=0.7)
title('Congruent Data')
xlabel('Task Time [s]')
ylabel('Normalized Frequency');
```

```python
# Statistical tests

import scipy.stats as stats

(t_statistic, p_value) = stats.ttest_rel(stroop_df['Congruent'], stroop_df['Incongruent'])
print "The t-statistic is %.3f" % t_statistic
print "The p-value is %.3f\n" % p_value

difference_df = stroop_df['Congruent'] - stroop_df['Incongruent']

cohens_d = difference_df.mean() / difference_df.std()
print "Cohen's d is %.3f" % cohens_d
```
